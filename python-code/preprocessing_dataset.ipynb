## Setting annotations up

```
ID = np.genfromtxt('id.csv', dtype=int)
d_left = np.genfromtxt('d_left.csv', delimiter=',')
d_right = np.genfromtxt('d_right.csv', delimiter=',')

df = pd.DataFrame([ID, d_left, d_right]).T
df.rename(columns={0: 'ID', 1: 'd_left', 2: 'd_right'}, inplace=True)
df.set_index('ID', inplace=True)
```

```
import re

filename_list = os.listdir('./dataset')
filename_list.sort()

ID_from_file = [filename.split('_')[0].lstrip('0') for filename in filename_list]
Side_from_file = [re.split('\_|\.', filename)[1] for filename in filename_list]

labels = [ df[f'd_{Side}'][int(ID)] for ID, Side in zip (ID_from_file, Side_from_file) ]
```

```
df_dataset = pd.DataFrame({'filename': filename_list,
                           'label': labels})
df_dataset.to_pickle('dataset.pkl')
```
